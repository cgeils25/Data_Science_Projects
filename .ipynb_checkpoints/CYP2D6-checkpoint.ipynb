# Filtering Data

Since IC50/AC50 is a reliable measure of activity, we'll isolate for those. This is as opposed to inhibition as a %, which is effectively some unknown point on a concentration/inhibition plot

```python
# creating new dataframe with only AC50 and IC50 with nM measurement
df_filtered = df[(df['Standard_Type'].isin(['IC50', 'AC50'])) & (df['Standard_Units'] == 'nM')]
df_filtered.reset_index(inplace=True)
df_filtered.drop('index', axis=1, inplace=True)
```

```python
# clearing original dataframe from memory
del df
```

## Investigating Standard Relation

I noticed that for the standard relation, some of the values were not "=", meaning it was not clear if the value given for AC50/IC50 was actually the value or just a bound. Here, I investigated the standard values (aka AC50 and IC50 values) to see if these are safe to include.

```python
df_filtered.Standard_Relation.value_counts()
```

```python
df_filtered.Standard_Value.describe()
```

```python
df_filtered[df_filtered['Standard_Relation'] == "'='"].Standard_Value.describe()
```

```python
df_filtered[df_filtered['Standard_Relation'] == "'>='"].Standard_Value.describe()
```

```python
df_filtered[df_filtered['Standard_Relation'] == "'<'"].Standard_Value.describe()
```

```python
df_filtered[df_filtered['Standard_Relation'] == "'<='"].Standard_Value.describe()
```

It's not clear whether these values are safe to include. I can't find good documentation, and its clear that the standard values for which the relationship is not an equality to not appear to be extreme. By this I mean that, for example, a standard relation of '>' might have a value of 2e+08, which would be at the upper end of the standard value distribution and potentially safe to include. As a result of this uncertainty, I've elected to exclude these values.

```python
df_filtered = df_filtered[df_filtered.Standard_Relation == "'='"]
df_filtered.Standard_Relation.value_counts()
```

# Extracting Model Data

I will extract the SMILES strings and AC50/IC50 values into a new dataframe for model building. My reasoning for not using the other experimental data given by ChEMBL is that I wanted to develop a model that could be used on enumerated molecules, which obviously would not have any experimental data as they may not have been tested in vitro/vivo

```python
# extracting smiles and standard value into dataframe for modeling
df_model = df_filtered[['Smiles', 'Standard_Value']]
df_model = df_model.rename(columns={'Standard_Value': 'IC50'})  # AC50 and IC50 are the same thing
df_model.reset_index(inplace=True)
df_model.drop('index', axis=1, inplace=True)
df_model.head()
```

```python
# clearing df_filtered from memory
del df_filtered
```

```python
# checking for null values
sns.heatmap(df_model.isna())
```

```python
df_model.Smiles.isna().value_counts()
```

```python
# removing null smiles and checking to make sure it worked
df_model.dropna(inplace=True)
df_model.Smiles.isna().value_counts()
```

## Visualizing Extreme Outliers

```python
# viewing distribution of IC50
df_model.describe()
```

```python
plt.boxplot(df_model.IC50)
```

There appears to be a few outliers, with highly significant outliers at ~1 billion. We can investigate this extreme value below:

```python
df_model[df_model['IC50'] == 1_000_000_000]
```

```python
# visualizing molecules with IC50 of 1 billion

from rdkit.Chem import AllChem
from IPython.display import display

# List of SMILES strings
smiles_list = df_model[df_model['IC50'] == 1_000_000_000].Smiles

# Convert SMILES to molecule objects and generate 2D coordinates
mols = [Chem.MolFromSmiles(smi) for smi in smiles_list]
for mol in mols:
    AllChem.Compute2DCoords(mol)

# Create a grid image of the molecules
grid_image = Draw.MolsToGridImage(mols, molsPerRow=2, subImgSize=(200, 200))

# Display the grid image (for Jupyter notebooks)
display(grid_image)
```

It's not clear why these molecules have such high IC50 values, other than the fact that they may not interact with CYP2D6 at all. I'll include these for now.
