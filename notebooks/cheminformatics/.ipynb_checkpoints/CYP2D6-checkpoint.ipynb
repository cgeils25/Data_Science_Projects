# Filtering Data

Since IC50/AC50 is a reliable measure of activity, we'll isolate for those. This is as opposed to inhibition as a %, which is effectively some unknown point on a concentration/inhibition plot

```python
# creating new dataframe with only AC50 and IC50 with nM measurement
df_filtered = df[(df['Standard_Type'].isin(['IC50'])) & (df['Standard_Units'] == 'nM')]
df_filtered.reset_index(inplace=True)
df_filtered.drop('index', axis=1, inplace=True)
```

It's not clear whether these values are safe to include. I can't find good documentation, and its clear that the standard values for which the relationship is not an equality to not appear to be extreme. By this I mean that, for example, a standard relation of '>' might have a value of 2e+08, which would be at the upper end of the standard value distribution and potentially safe to include. As a result of this uncertainty, I've elected to exclude these values.

```python
df_filtered = df_filtered[df_filtered.Standard_Relation == "'='"]
df_filtered.Standard_Relation.value_counts()
```

# Extracting Model Data

I will extract the SMILES strings and AC50/IC50 values into a new dataframe for model building. My reasoning for not using the other experimental data given by ChEMBL is that I wanted to develop a model that could be used on enumerated molecules, which obviously would not have any experimental data as they may not have been tested in vitro/vivo.

Also, I obviously did not want to include values like ligand efficiency, since it effectively has the IC50 value built into it

```python
# extracting smiles and standard value into dataframe for modeling
df_model = df_filtered[['Smiles', 'Standard_Value']]
df_model = df_model.rename(columns={'Standard_Value': 'IC50'})  # AC50 and IC50 are the same thing
df_model.reset_index(inplace=True)
df_model.drop('index', axis=1, inplace=True)
df_model.head()
```

```python
# clearing df_filtered from memory
del df_filtered
```

```python
from rdkit.Chem import Descriptors
```

```python
Descriptors.CalcMolDescriptors(Chem.MolFromSmiles('C1ccccC6'))
```

```python
from rdkit import RDLogger
```

```python
# creating a dictionary with all molecular descriptor data for smiles
import time
import sys, os
RDLogger.DisableLog('rdApp.*')  # suppresses a bunch of warnings

sys.stdout = open(os.devnull, 'w')
start = time.time()
smiles_list = df_model.Smiles
all_descriptors_list = []
for i, smiles in enumerate(smiles_list):
    this_smiles_descriptors = Descriptors.CalcMolDescriptors(Chem.MolFromSmiles(smiles))  # dict of descriptor_name: descriptor
    all_descriptors_list.append(this_smiles_descriptors.values())
    if i == 0:  # get descriptor names from first one
        all_descriptor_names = list(this_smiles_descriptors.keys())

all_descriptors_concatenated = list(zip(*all_descriptors_list))

all_descriptors = {}
for i, descriptor_name in enumerate(all_descriptor_names):
    all_descriptors[descriptor_name] = all_descriptors_concatenated[i]

del all_descriptors_list, all_descriptors_concatenated, all_descriptor_names

print(time.time() - start)

RDLogger.EnableLog('rdApp.*')  # re-enable rdkit logging
```

```python
# converting descriptors dictionary to dataframe
df_desc = pd.DataFrame(all_descriptors)
df_desc.head()
```

```python
# concatenating descriptors with original model dataframe
df_model = pd.concat([df_model, df_desc], axis=1)
```

```python
# deleting descriptor dataframe and dict
del all_descriptors, df_desc
```

```python
df_model.head()
```

```python
# searching for null values
sns.heatmap(df_model.isna())
```

```python
# identifying columns with null values, and the number of null values
for col in df_model:
    isna_bool = df_model[col].isna()
    if any(isna_bool):
        print(col, " ", sum(isna_bool)) 
```

```python
# because of the relatively small number of null values, I'll assume it's safe to drop these
df_model.dropna(inplace=True)
```

```python
# validating that null values were removed
for col in df_model:
    isna_bool = df_model[col].isna()
    if any(isna_bool):
        print(col, " ", sum(isna_bool)) 
# no output means it worked
```

# Modeling Using Descriptors

```python
df_model.head()
```

```python
"""
Classification of IC50 values:
0-150: strong inhibition
151-400: weak inhibition
400: little/no inhibition
"""


def get_activity(IC50_val):
    if IC50_val <= 150:
        return 'strong inhibition'
    elif IC50_val > 150 and IC50_val <= 600:
        return 'weak inhibition'
    elif IC50_val > 400:
        return 'little/no inhibition'
    else:
        print('error')
```

```python
# classifying acitivty as either active, moderately active, or inactive based on IC50
df_model['Activity'] = df_model['IC50'].apply(lambda x: get_activity(x))
df_model.head()
```

```python
df_model.Activity.value_counts()
```

```python
# visualizing the differences in some of the means in descriptors for the classes
thing = df_model.drop('Smiles', axis=1).groupby('Activity').mean()
for i, key in enumerate(thing):
    sns.barplot(thing[key])
    plt.show()
    plt.close()
    if i == 5:
        break
```

```python
len(df_model[df_model.IC50 == 0])
```

```python
from imblearn.ensemble import BalancedRandomForestClassifier
```

```python
model = BalancedRandomForestClassifier(bootstrap=True, replacement=True, sampling_strategy='all')
```

```python
X = np.array(df_model.drop(['IC50', 'Smiles', 'Activity'], axis=1))
y = np.array(df_model['Activity'])
```

```python
X.shape
```

```python
from sklearn.model_selection import train_test_split
x_train, x_test, y_train, y_test = train_test_split(X, y, test_size = 0.2)
```

```python
model.fit(x_train, y_train)
```

```python
y_out = model.predict(x_test)
```

```python
from sklearn.metrics import accuracy_score, confusion_matrix, classification_report
```

```python
# obtaining overall model accuracy
accuracy_score(y_test, y_out)
```

```python
# obtaining classification report for model
print(classification_report(y_test, y_out))
```

# Modeling Using Fingerprints

```python
df_fp = df_model[['Smiles', 'Activity']]
df_fp
```

```python
from rdkit.Chem import rdFingerprintGenerator
mfpgen = rdFingerprintGenerator.GetMorganGenerator(radius=3,fpSize=512)
```

```python
mol = Chem.MolFromSmiles(df_fp.iloc[5].Smiles)
np.array(mfpgen.GetFingerprint(mol))
```

```python
df_fp.iloc[1].Smiles
```

```python
def get_fingerprint(smiles):
    mol = Chem.MolFromSmiles(smiles)
    return np.array(mfpgen.GetFingerprint(mol))
```

```python
x = np.column_stack(df_model.Smiles.apply(lambda smi: get_fingerprint(smi))).T
y = np.array(df_fp['Activity'])
```

```python
x.shape
```

```python
# verifying I didn't mess up the data 
all(x[15] == get_fingerprint(df_model.Smiles[15]))
```

```python
x_train, x_test, y_train, y_test = train_test_split(x, y, test_size=0.2)
```

```python
model = BalancedRandomForestClassifier(bootstrap=True, replacement=True, sampling_strategy='all')
```

```python
model.fit(x_train, y_train)
```

```python
y_out = model.predict(x_test)
```

```python
# obtaining overall model accuracy
accuracy_score(y_test, y_out)
```

```python
# obtaining classification report for model
print(classification_report(y_test, y_out))
```

```python
visualize_confusion_matrix('Confusion Matrix for Balanced Random Forest Using Morgan Fingerprints', y_test, y_out)
```
